```python
# Examine the shape of the DataFrame
print("DataFrame Shape:", df.shape)

# Display the data types of each column
print("\nDataFrame Data Types:\n", df.dtypes)

# Get summary statistics for numerical columns
print("\nSummary Statistics:\n", df.describe())

# Check for missing values
print("\nMissing Values per column:\n", df.isnull().sum())

# Display the last few rows of the DataFrame
display(df.tail())

# Identify relevant columns for filtering
print("\nObservations on Columns:")
for col in df.columns:
    print(f"\nColumn: {col}")
    print(f"  Data Type: {df[col].dtype}")
    print(f"  Unique Values: {df[col].nunique()}")
    print(f"  Sample Values: {df[col].unique()[:5]}") # Display first 5 unique values
    if df[col].dtype == 'object':
      print(f"  Example value: {df[col].iloc[0]}")
```

## Data analysis

### Subtask:
Demonstrate various filtering techniques on the DataFrame.


**Reasoning**:
Apply different filtering techniques to the DataFrame as instructed, and display the results.



```python
# Single Column Value Filtering
df_technology = df[df['Category'] == 'Technology']
print("Filtering by Category == 'Technology'")
display(df_technology.head())
print(f"Shape of df_technology: {df_technology.shape}")

# Multiple Conditions Filtering
df_amount_supplier = df[(df['Invoice Amount ($)'] > 10000) & (df['Supplier Name'] == 'Supplier A')]
print("\nFiltering by Invoice Amount > 10000 and Supplier Name == 'Supplier A'")
display(df_amount_supplier.head())
print(f"Shape of df_amount_supplier: {df_amount_supplier.shape}")

# String Method Filtering
df_supplier_substring = df[df['Supplier Name'].str.contains('Supplier')]
print("\nFiltering by Supplier Name containing 'Supplier'")
display(df_supplier_substring.head())
print(f"Shape of df_supplier_substring: {df_supplier_substring.shape}")

# Numerical Range Filtering
df_amount_range = df[(df['Invoice Amount ($)'] >= 15000) & (df['Invoice Amount ($)'] <= 25000)]
print("\nFiltering by Invoice Amount between 15000 and 25000 (inclusive)")
display(df_amount_range.head())
print(f"Shape of df_amount_range: {df_amount_range.shape}")

# Date Filtering
try:
    df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
    start_date = pd.to_datetime('2025-01-01')
    end_date = pd.to_datetime('2025-01-05')
    df_date_range = df[(df['Invoice Date'] >= start_date) & (df['Invoice Date'] <= end_date)]
    print("\nFiltering by Invoice Date between 2025-01-01 and 2025-01-05")
    display(df_date_range.head())
    print(f"Shape of df_date_range: {df_date_range.shape}")
except Exception as e:
    print(f"\nDate filtering failed: {e}")
    df_date_range = None
```

```python
import pandas as pd
import random

# Generate 20 sample POs
po_numbers = [f"PO-{str(i).zfill(3)}" for i in range(1, 21)]
supplier_names = [f"Supplier {chr(65+i)}" for i in range(20)]

# Simulate realistic PO Amounts and Spent Amounts
po_amounts = [random.randint(20000, 100000) for _ in range(20)]
amounts_spent = []

for amount in po_amounts:
    # Randomly decide if PO is healthy or risky
    if random.random() < 0.3:  # 30% risky
        # Spend 90% or more of the amount
        spent = amount * random.uniform(0.91, 1.00)
    else:
        # Healthy POs: 40% to 85% spent
        spent = amount * random.uniform(0.40, 0.85)
    amounts_spent.append(round(spent, 2))

# Build DataFrame
df = pd.DataFrame({
    'PO Number': po_numbers,
    'Supplier Name': supplier_names,
    'PO Amount ($)': po_amounts,
    'Amount Spent ($)': amounts_spent
})

# Calculate Available Funds and %
df['Available Funds ($)'] = df['PO Amount ($)'] - df['Amount Spent ($)']
df['Available Funds (%)'] = (df['Available Funds ($)'] / df['PO Amount ($)']) * 100

# Flag Low Funds
df['Low Funds Flag'] = df['Available Funds (%)'] < 10

# Preview
print(df)

# Save report
df.to_csv('po_funds_monitoring_full_report.csv', index=False)
print("\nPO Funds Monitoring Report saved as 'po_funds_monitoring_full_report.csv'")
```

```python
import matplotlib.pyplot as plt

# Bar graph of Available Funds % by PO
plt.figure(figsize=(12,6))
plt.bar(df['PO Number'], df['Available Funds (%)'], color='skyblue')
plt.axhline(10, color='red', linestyle='--', label='10% Risk Threshold')
plt.title('PO Available Funds % Overview')
plt.xlabel('PO Number')
plt.ylabel('Available Funds (%)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Define a function to assign Risk Levels based on Available Funds (%)
def assign_risk_level(available_percent):
    if available_percent < 10:
        return 'High Risk'
    elif available_percent < 20:
        return 'Monitor'
    else:
        return 'Healthy'

# Apply the function to the DataFrame
df['Risk Level'] = df['Available Funds (%)'].apply(assign_risk_level)

# View updated DataFrame
print(df[['PO Number', 'Supplier Name', 'PO Amount ($)', 'Amount Spent ($)', 'Available Funds ($)', 'Available Funds (%)', 'Risk Level']])
```
